```python
import pandas as pd
import ast
```

```python
data = pd.read_csv('new.csv', low_memory=False)
```

```python
data['production_companies'] = data['production_companies'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else [])
```

```python
def eng(x):
    llsst=[]
    if x==[]:
        #print(1)
        llsst.append(0)
    else:
        m=x[0]
        #print(m['name'])
        
        llsst.append(m['name'])

    return llsst



data['company']=data['production_companies'].apply(lambda x: eng(x))
```

```python
data['production_countries'] = data['production_countries'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else [])
```

```python
import ast
import math
import pandas as pd
import numpy as np



def eng(x):
    llsst=[]
    try:
        if pd.isna(x):  
            llsst.append(0)
        elif not x:  
            llsst.append(0)
        else:
            a = ast.literal_eval(x)
            if a:
                llsst.append(a[0]['name'])
    except (SyntaxError, ValueError):
        llsst.append(0)
    return llsst

data['country'] = data['production_countries'].apply(lambda x: eng(x))
```
